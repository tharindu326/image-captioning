```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random
import matplotlib.gridspec as gridspec


def display_images_with_captions(data, num_images=5):
    for k, v in data.items():
        selected = random.sample(v, num_images)
        fig = plt.figure(figsize=(10, 2 * num_images))
        name = None
        if k == 'train':
            name = 'Train Data'
        if k == 'test':
            name = 'Test Data'
        if k == 'validation':
            name = 'Validation Data'
        fig.suptitle(name, fontsize=16)
        gs = gridspec.GridSpec(num_images, 2, width_ratios=[1, 2])

        for i, item in enumerate(selected):
            # Extract the image path and caption
            img_path = list(item.values())[0]['image_path']
            caption = list(item.values())[0]['caption']

            ax0 = plt.subplot(gs[i, 0])
            img = mpimg.imread(img_path)
            ax0.imshow(img)
            ax0.axis('off')

            # Display caption
            ax1 = plt.subplot(gs[i, 1])
            ax1.text(0.5, 0.5, caption, ha='center', va='center', fontsize=12, wrap=True,
                     bbox=dict(facecolor='white', alpha=0.5, edgecolor='gray', boxstyle='round,pad=0.5'))
            ax1.axis('off')

        plt.tight_layout()
        plt.show()
```

```python
import json

json_data_path = 'selected_dataset/selected_dataset_info.json'
data = json.load(open(json_data_path))

display_images_with_captions(data, num_images=5)
```
